```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Les donnees a visualiser
prix_ms = [39,50,57,69,94,100]
prix_apple = [93,112,104,144,170,172]
annee = [2015,2016,2017,2018,2019,2020]
```

```python
# Visualiser les donnees
plt.plot(annee, prix_ms, label='Prix Windows', c='g')
plt.plot(annee, prix_apple, label='Prix Apple', c='r')
plt.legend()
plt.xlabel('Annee')
plt.ylabel('Prix des ordinateurs')
plt.title('Evolution des Prix des ordinateurs durant les 5 dernieres annees')
```

```python
# Visualiser les donnees
# scatter - avec les nuages de points
plt.scatter(annee, prix_ms, label='Prix Windows', c='g')
plt.plot(annee, prix_apple, label='Prix Apple', c='r')
plt.legend()
plt.xlabel('Annee')
plt.ylabel('Prix des ordinateurs')
plt.title('Evolution des Prix des ordinateurs durant les 5 dernieres annees')
```

```python
# Visualiser les donnees dans 2 plans
# scatter - avec les nuages de points
plt.subplot(2,1,1) # 2 : nb de ligne - 1 : nb colonne - 1 : position
plt.scatter(annee, prix_ms, label='Prix Windows', c='g')

plt.subplot(2,1,2)
plt.plot(annee, prix_apple, label='Prix Apple', c='r')
```

```python
# Visualiser les donnees en diagramme de bande
plt.bar(annee, prix_ms, label='Prix Windows')
plt.bar(annee, prix_apple, label='Prix Apple', bottom=prix_ms)
plt.legend()
plt.xlabel('Annee')
plt.ylabel('Prix des ordinateurs')
plt.title('Evolution des Prix des ordinateurs durant les 5 dernieres annees')
```

```python
# Visualiser les donnees en diagramme de bande
index_annee = np.arange(len(annee))
width = 0.25

plt.bar(index_annee, prix_ms, label='Prix Windows', width=0.25)
plt.bar(index_annee+width, prix_apple, label='Prix Apple', width=0.25)
plt.legend()
plt.xlabel('Annee')
plt.ylabel('Prix des ordinateurs')
plt.title('Evolution des Prix des ordinateurs durant les 5 dernieres annees')
```

```python
# diagramme en Secteur - Pie Chart
produit = ['P1','P2','P3','P4']
vente = [10,20,20,50]
plt.pie(vente,labels=produit, autopct='%1.1f%%')
```
